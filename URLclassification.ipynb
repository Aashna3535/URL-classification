ML problem: Identifying URL is relevant or not.

For any textual data, all the categorical and textual data needs to be converted into numerical form before feeding to the machine learning algorithm, and that can be done by Word Embedding. It is a ML technique to convert a word into vectors. Some popular ways of doing that are one-Hot encoding, Bag of words, TF-IDF, Word2vec, FastText, etc. After that, the data set is ready to be applied to the ML algorithm. 

Steps:
1.   Loading of json files into data frames
2.   Combining those dataframes into one
3.   Shuffling it to prepare an unbiased data set
4.   Feature extraction
5.   Text Cleaning
6.   Tokenization
7.   Making the bag of words via sparse matrix
8.   Splitting of data set
9.   Fitting the training data to random forest classifier
10.  Predicting the test data
11.  Generating classification report

```python
#Splitting the feature iab_cat to iab and conf
def getvalues(dict): 
    return dict.values()

i=0;
j=0;
iab_cat_conf = []
iab_cat_iab = []

while i< len(finaldatacopy.iab_cat):
    for j in range(0,1):
        iab_cat_conf.append((finaldatacopy.iab_cat[i][j]).get('confident'))
        iab_cat_iab.append((finaldatacopy.iab_cat[i][j]).get('iab'))
        j+=1
    i+=1
    
iab_cat_conf
#print(iab_cat_iab)
```

```python
#Assigning numerics to feature languages according to values
df_lang = pd.DataFrame(finaldatacopy['language'].replace(['de','en','es','fr','it','pt'],[0,1,2,3,4,5]))
df_lang
```

```python
#Adding updated languages feature to the training data
finaldatacopy['language'] = df_lang
finaldatacopy
```

```python
#Updating feature hashtags based on number of keywords presents
ls = []
for i in range(len(finaldatacopy)): 
    ls.append(len(finaldatacopy.hashtags[i]))
    
df_hashtags = pd.DataFrame(ls)
df_hashtags
```

```python
#Adding updated feature hashtags to the training data set
finaldatacopy['hashtags'] = df_hashtags
finaldatacopy
```

```python
#Removing previous iab_cat feature
finaldatacopy = finaldatacopy.drop(['iab_cat'], axis = 1)
finaldatacopy
```

```python
import re   
import nltk  
  
nltk.download('stopwords') 
  
from nltk.corpus import stopwords  
from nltk.stem.porter import PorterStemmer 
 
 
corpus = []  
```

```python
# 1669 (text) rows to clean 
for i in range(0, 1669):  
      
    # column : "text", row ith 
    text = re.sub('[^a-zA-Z]', ' ', finaldatacopy['text'][i])  
      
    # convert all cases to lower cases 
    text = text.lower()  
      
    # split to array(default delimiter is " ") 
    text = text.split()  
      
    # creating PorterStemmer object to 
    # take main stem of each word 
    ps = PorterStemmer()  
      
    # loop for stemming each word 
    # in string array at ith row     
    text = [ps.stem(word) for word in text 
                if not word in set(stopwords.words('english'))]  
                  
    # rejoin all string array elements 
    # to create back into a string 
    text = ' '.join(text)   
      
    # append each string to create 
    # array of clean text  
    corpus.append(text)  
```

```python
finaldatacopy['text'] = corpus
finaldatacopy['text']
```

```python
# import the existing word and sentence tokenizing  
# libraries 
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk
nltk.download('punkt')
```

```python
for i in range(0,len(finaldatacopy)):
  print(sent_tokenize(finaldatacopy['text'][i])) 
  print(word_tokenize(finaldatacopy['text'][i]))
```

```python
# Creating the Bag of Words model 
from sklearn.feature_extraction.text import CountVectorizer 
  
# To extract max 1500 feature. 
# "max_features" is attribute to 
# experiment with to get better results 
cv = CountVectorizer(max_features = 1500)  
  
# txt contains corpus 
txt = cv.fit_transform(corpus).toarray()  
txt  
```

```python
cor = []
# 1669 (url) rows to clean 
for i in range(0, 1669):  
      
    # column : "text", row ith 
    text = re.sub('[^a-zA-Z]', ' ', finaldatacopy['url'][i])  
      
    # convert all cases to lower cases 
    text = text.lower()  
      
    # split to array(default delimiter is " ") 
    text = text.split()  
      
    # creating PorterStemmer object to 
    # take main stem of each word 
    ps = PorterStemmer()  
      
    # loop for stemming each word 
    # in string array at ith row     
    text = [ps.stem(word) for word in text 
                if not word in set(stopwords.words('english'))]  
                  
    # rejoin all string array elements 
    # to create back into a string 
    text = ' '.join(text)   
      
    # append each string to create 
    # array of clean text  
    cor.append(text) 
```

```python
finaldatacopy['url'] = cor
finaldatacopy['url']
```

```python
for i in range(0,len(finaldatacopy)):
  print(sent_tokenize(finaldatacopy['url'][i])) 
  print(word_tokenize(finaldatacopy['url'][i]))
```

```python
 
# urls contains corpus  
urls = cv.fit_transform(cor).toarray()  
urls  
```

```python
#Replacing NAN values to 0 and identifying categories of iab_cat_iab
finaldatacopy['iab_cat_iab'] = finaldatacopy['iab_cat_iab'].replace(np.nan, 0)
finaldatacopy['iab_cat_iab'].value_counts()
```

```python
#Printing distinct categories
s = set(finaldatacopy['iab_cat_iab'])
s
```

```python
#Creating dictionary of category and its corresponding value
tl = list(finaldatacopy['iab_cat_iab'])
class my_dictionary(dict): 
  
    # __init__ function 
    def __init__(self): 
        self = dict() 
          
    # Function to add key:value 
    def add(self, key, value): 
        self[key] = value 
        
dict_obj = my_dictionary() 

for i in range(0,166):
    dict_obj.add(tl[i], i)
    
dict_obj
```

```python
#Replacing iab_cat_values to its corresponding category
finaldatacopy['iab_cat_iab'] = finaldatacopy['iab_cat_iab'].replace(s,list(range(0, 167)))
finaldatacopy['iab_cat_iab']
```

```python
#Creating target variable
y = pd.DataFrame(finaldatacopy['category'])
y
```

```python
#Dropping target variable and text fields
finaldatacopy = finaldatacopy.drop(['text','url','category'],axis = 1)
finaldatacopy
```

```python
#Converting the dataframe to array
arr = finaldatacopy.to_numpy()
arr
```

```python
#Concatenating arrays of text, url and other remaining features
x = np.concatenate([txt, urls,arr], axis =1)
x
```

```python
# Splitting the dataset into the Training set and Test set 
import sklearn
from sklearn.model_selection import train_test_split 
  

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20) 
```

```python
# Fitting Random Forest Classification to the Training set 
from sklearn.ensemble import RandomForestClassifier 

# n_estimators can be said as number of trees

model = RandomForestClassifier(n_estimators = 501, 
							criterion = 'entropy') 
							
model.fit(x_train, y_train) 
```

```python
# Predicting the Test set results 
y_pred = model.predict(x_test) 
y_pred 
```

```python
# Printing classification report 
 
from sklearn.metrics import classification_report

y_test_array = y_test.to_numpy()
print(classification_report(y_test_array, y_pred))
```
